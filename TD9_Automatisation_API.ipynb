```python
def Get_list_crypto():
    crypto_lst=[]
    resp=rq.get('https://api.binance.com/api/v3/exchangeInfo')
    data_list=resp.json()
    for i in data_list['symbols']:
        crypto_lst.append(i['baseAsset'])
        crypto_lst=list(dict.fromkeys(crypto_lst)) #suppresion des doublons
    return crypto_lst
```

• Get order book for an asset

```python
def getDepth(direction='asks', pair='BTCUSDT'):
    url = f"https://api.binance.com/api/v3/depth?symbol={pair}" 
    resp = rq.get(url)
    data = resp.json()
    depth = data[direction]
    return depth
```

Order Book

```python
def getOrderBook(pair='BTCUSDT'):
    asks=getDepth(direction='asks',pair=pair)
    bids=getDepth(direction='bids',pair=pair)
    return asks,bids
```

• Create a function to read agregated trading data (candles)  
def refreshDataCandle(pair = 'BTCUSD', duration = '5m’)

data structure for a candle:  
[  
  [    
          1499040000000,      // Kline open time  
          "0.01634790",       // Open price  
          "0.80000000",       // High price  
          "0.01575800",       // Low price  
          "0.01577100",       // Close price  
          "148976.11427815",  // Volume  
          1499644799999,      // Kline Close time  
          "2434.19055334",    // Quote asset volume  
          308,                // Number of trades  
          "1756.87402397",    // Taker buy base asset volume  
          "28.46694368",      // Taker buy quote asset volume  
          "0"                 // Unused field, ignore.  
    ]  
]  

```python
#Version n°1 refreshDataCandle

def refreshDataCandle(pair='BTCUSDT', duration='5m'):
    lst = []
    Liste_attribut = ['Kline open time', 'Open price', 'High price', 'Low price',
                      'Close price', 'Volume', 'Kline Close time', 'Quote asset volume',
                      'Number of trades', 'Taker buy base asset volume', 'Taker buy quote asset volume',
                      'Unused field ignore']

    resp = rq.get(f"https://api.binance.com/api/v3/klines?symbol={pair}&interval={duration}")
    data = resp.json()

    # Vérification de la présence d'un message d'erreur dans la réponse
    if 'code' in data or 'msg' in data:
        print(f"Erreur de l'API: {data}")
        return []

    for i in data:
        try:
            float_data = [float(x) if j not in [0, 6] else int(x) for j, x in enumerate(i[:8])]
            dictData = zip(Liste_attribut, float_data)
            dictData = dict(dictData)
            lst.append(dictData)
        except ValueError as e:
            print(f"Erreur de conversion: {e}, dans la donnée: {i}")

    return lst

```

```python
#take the last maj
def getIndex(pair):
    conn = sqlite3.connect('BlockProg_td9.db')
    cursor = conn.cursor()
    Id=0
    sql_request = f"SELECT MAX(Id) FROM {pair}"
    Id_ex=cursor.execute(sql_request)
    for i in Id_ex:
        Id=i
    conn.close()
    return Id


#avoir last date maj
def getLastMaj(pair):
    conn = sqlite3.connect('BlockProg_td9.db')
    cursor = conn.cursor()
    last_date=0
    sql_request = f"SELECT MAX(`date`) FROM {pair}"
    Id_ex=cursor.execute(sql_request)
    for i in Id_ex:
        last_date=i
    conn.close()
    return last_date

#insert data in db
def StoreNewCandle(pair, List):
    conn = sqlite3.connect('BlockProg_td9.db')
    cursor = conn.cursor()
    
    index=getIndex(pair)[0]
    if index==None:
        index=0
    else:
        index=index+1
    
    for i in List:
        sql=f"INSERT INTO {pair} (Id,`date`,high,low,open,close,volume,quotevolume) VALUES ({index},'{i['Kline open time']}',{i['High price']},{i['Low price']},{i['Open price']},{i['Close price']},{i['Volume']},{i['Quote asset volume']})"
        cursor.execute(sql)
        index+=1
        conn.commit()
    conn.close()
    return True
```

```python
def refreshData(pair='BTCUSDT'):
    resp = rq.get(f"https://api.binance.com/api/v3/ticker/price?symbol={pair}")
    data = resp.json()
    if resp.status_code != 200:
        print(f"Erreur de l'API: {data}")
        return None
    return data
```

```python
def createTable(pair):
    try:
        connexion =sqlite3.connect("BlockProg_td9.db")
        cursor=connexion.cursor()
        cursor.execute(f'''
            CREATE TABLE {pair} (
                id INTEGER PRIMARY KEY,
                DATE INT,
                HIGH REAL,
                LOW REAL,
                OPEN REAL,
                CLOSE REAL,
                VOLUME REAL,
                QUOTEVOLUME REAL
            )

        ''')
        connexion.commit()
        connexion.close()
        return True
    except sqlite3.Error as e:
        print(f"Une erreur s'est produite lors de la création de la table: {e}")
        return False
```

```python
#Store data in Db
def storeCandleData(pair,storeCandleData):
    try:
        connexion = sqlite3.connect("BlockProg_td9.db")
        cursor= connexion.cursor()

        query=f''' INSERT INTO {pair} (
            DATE,
            HIGH,
            LOW,
            OPEN,
            CLOSE,
            VOLUME,
            QUOTEVOLUME)

            VALUES(
                ?,
                ?,
                ?,
                ?,
                ?,
                ?,
                ?
        )'''

        data_to_insert=(
            storeCandleData['Kline open time'], storeCandleData['High price'],
            storeCandleData['Low price'], storeCandleData['Open price'],
            storeCandleData['Close price'], storeCandleData['Volume'],
            storeCandleData['Quote asset volume'])
        
        cursor.execute(query,data_to_insert)

        connexion.commit()
        connexion.close()

        return True

    except sqlite3.Error as e:
        print(f"Une erreur s'est produite lors de l'insertion des données: {e}")
        return False
    finally:
        if connexion is not None:
            connexion.close()
```
